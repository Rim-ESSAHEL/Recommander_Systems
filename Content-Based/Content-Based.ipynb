```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
import string
import pandas as pd
import pickle 
```

```python
df = pd.read_csv('Cat_data.csv')
df.drop(columns=['Unnamed: 0'],inplace=True)
df
```

```python
list_stopwords = set(stopwords.words('english') + list(punctuation))

def lower(df,column):
    df[column] = df[column].astype(str).str.lower()
    
def tokenize(df,column):
    df[column] = df[column].apply(word_tokenize)
    
def rmv_stop_words(df,column):
    df[column] = df[column].apply(lambda x: [word for word in x if word not in list_stopwords])
    
def set_(df,column):
    df[column] = df[column].apply(lambda x : list(set(x)))
    
```

```python
columns = ['adult','genres','overview']
```

```python
for col in columns:
    print(col)
    lower(df,col)
    tokenize(df,col)
    rmv_stop_words(df,col)
    set_(df,col)
```

```python
with open('matrix_vocab.pickle', 'rb') as handle:
    matrix_vocab = pickle.load(handle)
#matrix_vocab
df = pd.read_csv('content.csv')
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head(44)
```

```python
import gensim

wv = gensim.models.KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)
```

```python
from tqdm import tqdm

def content_recommander(movie_title):
    matrix_title_vocab = []
    for list_ in df[df['original_title'] == movie_title].to_numpy():
        list_[0] = [word for word in list_[0] if word in wv.key_to_index]
        list_[1] = [word for word in list_[1] if word in wv.key_to_index]
        list_[4] = [word for word in list_[4] if word in wv.key_to_index]
        matrix_title_vocab.append(list_)

    matrix_similarity = []
    pbar = matrix_vocab
    for list1 in pbar:
        for list2 in matrix_title_vocab:  
            try:
                score_adult = wv.n_similarity(list1[0], list2[0])
            except:
                score_adult = 0
            try:
                score_genre = wv.n_similarity(list1[1], list2[1])
            except:
                score_genre = 0
            try:
                score_overview = wv.n_similarity(list1[4], list2[4])
            except:
                score_overview = 0
            if ((list1[4] != list2[4]) ):
                matrix_similarity.append([ list1[3], list2[3], list1[0], list2[0],list1[1], list2[1],list1[4], list2[4], score_adult, score_genre,score_overview])

    dfsimilarity = pd.DataFrame(matrix_similarity, columns = ['recommendation','movie_request','adult','adult_request','genre','genre_request','overview','overview_request','score_adult', 'score_genre','score_overview'])
    dfsimilarity['final_score'] = dfsimilarity['score_adult'] + dfsimilarity['score_genre']+ dfsimilarity['score_overview']
    return (dfsimilarity.sort_values(by=['final_score',  'score_genre', 'score_overview'], ascending=False).head(10))
```

```python
rec = content_recommander('Richard III')
```
